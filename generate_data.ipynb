```python
from tqdm import tqdm
```

```python
p_arr = []
x_arr = []
y_arr = []
z_arr = []

sum_arithmetic_arr = []
sum_arithmetic_sign_arr = []
sum_modulo_arr = []

sub_arithmetic_arr = []
sub_arithmetic_sign_arr = []
sub_modulo_arr = []

mul_arithmetic_arr = []
mul_modulo_arr = []
quot_modulo_arr = []
quot_modulo_sign_arr = []
mod_modulo_arr = []
inverse_modulo_arr = []

v_arr = []
r_arr = []
r2_arr = []

x_montgomery_arr = []
y_montgomery_arr = []

mul_montgomery_arr = []
```

```python
R = 2**128
begin = 2**127
end = R - 1
```

```python
for i in tqdm(range(100000)):
    p = random_prime(2**128 - 1, True, 2**127)
    x = randint(begin, end)
    y = randint(begin, end)
    z = randint(2**128, 2**256 - 1)
    x %= p
    y %= p
    
    p_arr.append(hex(p)[2:])
    x_arr.append(hex(x)[2:])
    y_arr.append(hex(y)[2:])
    z_arr.append(hex(z)[2:])
    
    sum_arithmetic = x + y
    if sum_arithmetic >= R:
        sum_arithmetic_arr.append(hex(sum_arithmetic)[3:])
        sum_arithmetic_sign_arr.append(1)
    else:
        sum_arithmetic_arr.append(hex(sum_arithmetic)[2:])
        sum_arithmetic_sign_arr.append(0)
    
    sum_modulo = (x + y) % p
    sum_modulo_arr.append(hex(sum_modulo)[2:])
    
    sub_arithmetic = x - y
    if sub_arithmetic < 0:
        sub_arithmetic_arr.append(hex(R + sub_arithmetic)[2:])
        sub_arithmetic_sign_arr.append(1)
    else:
        sub_arithmetic_arr.append(hex(sub_arithmetic)[2:])
        sub_arithmetic_sign_arr.append(0)
    
    sub_modulo = (x - y) % p
    sub_modulo_arr.append(hex(sub_modulo)[2:])
    
    mul_arithmetic = x * y
    mul_arithmetic_arr.append(hex(mul_arithmetic)[2:])
    
    mul_modulo = mul_arithmetic % p
    mul_modulo_arr.append(hex(mul_modulo)[2:])
    
    quot_modulo = z // p
    
    if quot_modulo >= R:
        quot_modulo_arr.append(hex(quot_modulo - R)[2:])
        quot_modulo_sign_arr.append(1)
    else:
        quot_modulo_arr.append(hex(quot_modulo)[2:])
        quot_modulo_sign_arr.append(0)
        
    mod_modulo = z % p
    mod_modulo_arr.append(hex(mod_modulo)[2:])
    
    inverse_modulo = inverse_mod(x, p)
    inverse_modulo_arr.append(hex(inverse_modulo)[2:])
    
    v = R - inverse_mod(p, R)
    v_arr.append(hex(v)[2:])
    
    r = R % p
    r_arr.append(hex(r)[2:])
    
    r2 = (R * R) % p
    r2_arr.append(hex(r2)[2:])
    
    x_montgomery = (x * R) % p
    x_montgomery_arr.append(hex(x_montgomery)[2:])
    
    y_montgomery = (y * R) % p
    y_montgomery_arr.append(hex(y_montgomery)[2:])
    
    mul_montgomery = (x * y * R) % p
    mul_montgomery_arr.append(hex(mul_montgomery)[2:])
```

```python
with open('data.json', 'w') as f:
    f.write('[')
    for i in range(100000):
        f.write('{')
        f.write('"p":"{}",'.format(p_arr[i]))
        f.write('"x":"{}",'.format(x_arr[i]))
        f.write('"y":"{}",'.format(y_arr[i]))
        f.write('"z":"{}",'.format(z_arr[i]))
        
        f.write('"sum_arithmetic":"{}",'.format(sum_arithmetic_arr[i]))
        f.write('"sum_arithmetic_sign":{},'.format(sum_arithmetic_sign_arr[i]))
        f.write('"sum_modulo":"{}",'.format(sum_modulo_arr[i]))
        
        f.write('"sub_arithmetic":"{}",'.format(sub_arithmetic_arr[i]))
        f.write('"sub_arithmetic_sign":{},'.format(sub_arithmetic_sign_arr[i]))
        f.write('"sub_modulo":"{}",'.format(sub_modulo_arr[i]))
        
        f.write('"mul_arithmetic":"{}",'.format(mul_arithmetic_arr[i]))
        f.write('"mul_modulo":"{}",'.format(mul_modulo_arr[i]))
        f.write('"quot_modulo":"{}",'.format(quot_modulo_arr[i]))
        f.write('"quot_modulo_sign":{},'.format(quot_modulo_sign_arr[i]))
        f.write('"mod_modulo":"{}",'.format(mod_modulo_arr[i]))
        f.write('"inverse_modulo":"{}",'.format(inverse_modulo_arr[i]))
        
        f.write('"v":"{}",'.format(v_arr[i]))
        f.write('"r":"{}",'.format(r_arr[i]))
        f.write('"r2":"{}",'.format(r2_arr[i]))
        
        f.write('"x_montgomery":"{}",'.format(x_montgomery_arr[i]))
        f.write('"y_montgomery":"{}",'.format(y_montgomery_arr[i]))
        
        f.write('"mul_montgomery":"{}"'.format(mul_montgomery_arr[i]))
        f.write('}')
        if i != 99999:
            f.write(',')
    f.write(']')
```
